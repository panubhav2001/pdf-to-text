```python
from pdf2image import convert_from_path
import pytesseract

def extract_text_from_pdf(pdf_path):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path)

    # Initialize a dictionary to store the text of each page
    page_texts = {}

    # Loop through images and apply OCR
    for i, image in enumerate(images):
        # Apply OCR to the image
        text = pytesseract.image_to_string(image)
        
        # Split the text into lines
        lines = text.splitlines()
        
        # Skip the first 7 lines and the last 3 lines
        if len(lines) > 10:  # Ensure there are enough lines to skip
            lines = lines[7:-3]
        else:
            lines = []
        
        # Join the remaining lines back into a single string
        processed_text = "\n".join(lines)
        
        # Optionally, store the text in a dictionary with the page number as the key
        page_texts[f'page_{i+1}'] = processed_text

    return page_texts

# Example usage:
pdf_path = 'Sample Problem.pdf'
extracted_texts = extract_text_from_pdf(pdf_path)

# Accessing the text of the first page
print(extracted_texts['page_1'])
```

```python
import re
import pandas as pd

def extract_voter_info(text):
    # Refined regex patterns
    name_pattern = r"Name\s*[:\-\*]\s*([A-Za-z ]+)(?=\s*Husband's Name|Father's Name|Name|House Number|Age|Gender|Photo|$)"
    relative_name_pattern = r"(?:Husband's Name|Father's Name)\s*[:\-\*]\s*([A-Za-z ]+)(?=\s*House Number|Photo|Age|Gender|$)"
    relation_type_pattern = r"(Husband|Father)'s Name"
    age_pattern = r"Age\s*[:\-\+]\s*(\d+)"
    gender_pattern = r"Gender\s*[:\-\+]\s*(Male|Female)"
    house_no_pattern = r"House Number\s*[:\-\+]\s*(\d+|Photo|Mangalwar|6a|7|0)?"
    epic_no_pattern = r"([A-Z0-9]{10})"

    # Extracting the information
    voter_names = re.findall(name_pattern, text)
    relative_names = re.findall(relative_name_pattern, text)
    relation_types = re.findall(relation_type_pattern, text)
    ages = re.findall(age_pattern, text)
    genders = re.findall(gender_pattern, text)
    house_nos = re.findall(house_no_pattern, text)
    epic_nos = re.findall(epic_no_pattern, text)

    # Ensure all lists have the same length
    max_len = max(len(voter_names), len(relative_names), len(relation_types), len(ages), len(genders), len(house_nos), len(epic_nos))

    voter_names += [''] * (max_len - len(voter_names))
    relative_names += [''] * (max_len - len(relative_names))
    relation_types += [''] * (max_len - len(relation_types))
    ages += [''] * (max_len - len(ages))
    genders += [''] * (max_len - len(genders))
    house_nos += [''] * (max_len - len(house_nos))
    epic_nos += [''] * (max_len - len(epic_nos))

    # Create a DataFrame
    data = {
        "Voter Full Name": voter_names,
        "Relative's Name": relative_names,
        "Relation Type": relation_types,
        "Age": ages,
        "Gender": genders,
        "House No": house_nos,
        "EPIC No": epic_nos,
    }

    df = pd.DataFrame(data)
    return df
# Calling the function
df1 = extract_voter_info(extracted_texts['page_1'])
df2 = extract_voter_info(extracted_texts['page_2'])
```

```python
# Concat DataFrames created from pdf pages
df = pd.concat([df1, df2],axis=0)
```

```python
# Exporting the DataFrame to Excel Workbook
df.to_excel('Extractedinfo.xlsx')
```
